# Asset valuation in the Mehra&ndash;Prescott economy

Mehra and Prescott (1985) proposed an economy with underlying uncertainty
described by a two-state Markov chain $x_{t}$ with transition matrix

\begin{equation*}
\mathbf{P}=\left[
\begin{array}{cc}
\phi  & 1-\phi  \\
1-\phi  & \phi
\end{array}%
\right] .
\end{equation*}

Consumption growth is modeled as a stationary process

\begin{equation*}
\frac{C_{t+1}}{C_{t}}=\exp \left( g_{C}\left( x_{t},x_{t+1}\right) \right)
\end{equation*}

represented by the matrix $\Gamma _{C}$ with elements

\begin{equation*}
\left[ \Gamma _{C}\right] _{ij}=\exp \left( g_{C}\left(
x_{t}=e_{i},x_{t+1}=e_{j}\right) \right) .
\end{equation*}

Preferences are of the CRRA\ form with time preference $\beta $ and relative
risk aversion $\gamma $, implying the SDF of the form

\begin{equation*}
\frac{S_{t+1}}{S_{t}}=\beta \left( \frac{C_{t+1}}{C_{t}}\right) ^{-\gamma
}=\beta \exp \left( -\gamma g_{C}\left( x_{t},x_{t+1}\right) \right)
\end{equation*}

represented by the matrix $\Gamma _{S}$ with elements

\begin{equation*}
\left[ \Gamma _{S}\right] _{ij}=\beta \exp \left( -\gamma g_{C}\left(
x_{t}=e_{i},x_{t+1}=e_{j}\right) \right) =\beta \left[ \Gamma _{C}\right]
_{ij}^{-\gamma }
\end{equation*}

where the $-\gamma $ power applies elementwise.

Mehra and Precott (1985) assume that the consumption growth matrix is given by

\begin{equation*}
\Gamma _{C}=\left[
\begin{array}{cc}
1+\mu +\delta  & 1+\mu -\delta  \\
1+\mu +\delta  & 1+\mu -\delta
\end{array}%
\right]
\end{equation*}

so that state $e_{1}$ is the high-growth state, state $e_{2}$ is the
low-growth state.

## Asset prices and returns

We are interested in pricing an infinite horizon asset that pays a cash flow stream $G_t$ with growth rate

\begin{equation*}
\frac{G_{t+1}}{G_{t}}=\exp \left( g_{G}\left( x_{t},x_{t+1}\right) \right)
\end{equation*}

represented by the matrix $\Gamma _{G}$ with elements

\begin{equation*}
\left[ \Gamma _{G}\right] _{ij}=\exp \left( g_{G}\left(
x_{t}=e_{i},x_{t+1}=e_{j}\right) \right) .
\end{equation*}

The valuation equation that yields price $Q_t$ of this asset is given by

$$Q_t = E_t \left[ \frac{S_{t+1}}{S_t} \left( G_{t+1} + Q_{t+1} \right) \right]$$

Since the one-period SDF and one-period cash-flow growth are stationary (they are functions of $x_t$ and $x_{t+1}$ only, which is a stationary process), then we can expect the price-dividend ratio $Q_t/G_t$ to be stationary as well. To see this, rewrite the valuation equation by dividing by $G_t$ as

\begin{equation}
\frac{Q_{t}}{G_{t}}=E_{t}\left[ \frac{S_{t+1}}{S_{t}}\left( \frac{Q_{t+1}%
}{G_{t+1}}+1\right) \frac{G_{t+1}}{G_{t}}\right].
\end{equation}

When we conjecture that $Q_{t+1}/G_{t+1}$ is a stationary function of the state $x_{t+1}$, then all terms inside of the expectations only depend on $x_t$ and $x_{t+1}$, and the expectation operator integrates over $x_{t+1}$. The Markov chain assumption then implies that the price-dividend ratio is indeed stationary:

\begin{equation*}
\frac{Q_{t}}{G_{t}}=q\left( x_{t}\right).
\end{equation*}

In the case when the cash flow $G_t$ is the consumption process itself, then $Q_t/C_t$ is the wealth-consumption ratio.

We can then express the return on the consumption claim (or 'equity' return) as

\begin{equation*}
R_{t+1}^{c}=\frac{Q_{t+1}+C_{t+1}}{Q_{t}}=\frac{q\left( x_{t+1}\right) +1}{%
q\left( x_{t}\right) }\frac{C_{t+1}}{C_{t}}
\end{equation*}

and the risk-free rate

\begin{equation*}
R_{t+1}^{f}=\left( E_{t}\left[ \frac{S_{t+1}}{S_{t}}\right] \right)
^{-1}.
\end{equation*}

## Matrix implementation

Since the state follows an $n$-state Markov chain, stationary functions of the state $x_t$ can be represented as vectors of length $n$. Denote the $\mathbf{q}$ the vector of state-dependent price-dividend ratios

\begin{equation*}
\left[ \mathbf{q}\right] _{i}=q\left( x_{t}=e_{i}\right) .
\end{equation*}

Using the matrix structure, the recursive valuation equation can be expressed as
\begin{equation*}
\left[ \mathbf{q}\right] _{i}=q\left( x_{t}=e_{i}\right) =\sum_{j=1}^{n}%
\left[ \mathbf{P}\right] _{ij}\left[ \Gamma _{S}\right] _{ij}\left[
\Gamma _{G}\right] _{ij}\left(\left[ \mathbf{q}\right] _{j} + 1 \right).
\end{equation*}

This can be written in compact form as

\begin{equation*}
\mathbf{q}=\left( \mathbf{P}\ast \Gamma _{S}\ast \Gamma _{G}\right)
\left( \mathbf{q}+\mathbf{1}\right)
\end{equation*}

where $\ast$ is the symbol for elementwise multiplication and $\mathbf{1}$ is a vector of ones. For convenience, denote

\begin{equation*}
\mathbf{M}=\left( \mathbf{P}\ast \Gamma _{S}\ast \Gamma _{G}\right).
\end{equation*}

The matrix $\mathbf{M}$ is also called the <b>valuation operator</b> because it takes next-period cash flow and produces this period value by properly discounting with marginal rate of substitution and probabilities of alternative future states.

We now want to solve for $\mathbf{q}$. In principle, we could iterate on the equation

\begin{equation*}
\mathbf{q_{n+1}}=\mathbf{M}
\left( \mathbf{q}_{n}+\mathbf{1}\right)
\end{equation*}

from a given initial value $\mathbf{q}_0$. As long as the contraction mapping property holds, there is a unique value $\mathbf{q}$ that solves the equation

\begin{equation*}
\mathbf{q}=\mathbf{M} \left( \mathbf{q}+\mathbf{1}\right)
\end{equation*}

and the iterative procedure based on successive approximations will converge to the value $\mathbf{q}$ from any initial $\mathbf{q}_0$.

The necessary and sufficient condition for this to be true is that the matrix $\mathbf{M}$ has to have all eigenvalues strictly inside the unit circle (see Boroviƒçka and Stachurski (2021) for details).

Intuitively, eigenvalues inside the unit circle mean that the valuation matrix $\mathbf{M}$ properly contracts the right-hand side vector $\mathbf{q} + \mathbf{1}$, making its magnitude 'smaller' in a proper sense, so that the iterations do not explode.

Rather than using successive approximations, we can solve the valuation equation directly by matrix inversion:

$$ \mathbf{q} = \left(\mathbf{I} - \mathbf{M}\right)^{-1} \mathbf{M} \mathbf{1}.$$

The conditions for this operations to be valid are the same as for the application of successive approximations, eigenvalues of $\mathbf{M}$ must be strictly inside the unit circle.

Following Mehra and Prescott (1985), we take the risky cash flow to be consumption itself, so that $G_t = C_t$ and thus $\Gamma_G = \Gamma_C$.

With the value of $q$ at hand, we can recover state-dependent returns, the risk-free rate, and expected returns in matrix form.

The return on the risky asset between $t$ and $t+1$ then is

\begin{equation*}
R_{t+1}^{c}=\frac{Q_{t+1}+C_{t+1}}{Q_{t}}=\frac{q\left( x_{t+1}\right) +1}{%
q\left( x_{t}\right) }\frac{C_{t+1}}{C_{t}}\qquad \implies \qquad \mathbf{R}%
_{ij}^{c}=\frac{\mathbf{q}_{j}+1}{\mathbf{q}_{i}}\left[ \Gamma _{C}\right]
_{ij}
\end{equation*}

The return depends both on the current state $x_t = e_i$ and next-period state $x_{t+1} = e_j$.

The risk-free rate is the return on an asset that pays one unit of consumption next period:

\begin{equation*}
R_{t+1}^{f}=\left( E_{t}\left[ \frac{S_{t+1}}{S_{t}}\right] \right)
^{-1}\qquad \implies \qquad \mathbf{R}_{ij}^{f}=\left( \left[ \mathbf{P}%
\ast \Gamma _{S}\right] _{i\cdot }\mathbf{1}\right) ^{-1}.
\end{equation*}

This risk-free rate depends on the current state $x_t = e_i$ but not on the next-period state $x_{t+1} = e_j$. This is the sense in which the risk-free rate is risk-free: risk-free rates can vary over time but at every date, they produce a guaranteed return that does not depend on the realization of the next-period state.

The conditional equity premium is the difference between expectations of the
two returns:

\begin{equation*}
E_{t}\left[ R_{t+1}^{c}-R_{t+1}^{f}|x_{t}=e_{i}\right] =\sum_{j=1}^{2}\left(
\mathbf{R}_{ij}^{c}-\mathbf{R}_{ij}^{f}\right) \mathbf{P}_{ij}.
\end{equation*}

Risk-premia will also generally depend on the current state $x_t = e_i$.

We are also interested in the unconditional premium, under the unconditional distribution $\pi $,

\begin{equation*}
E\left[ R_{t+1}^{c}-R_{t+1}^{f}\right] =\sum_{i=1}^{2}\pi
_{i}\sum_{j=1}^{2}\left( \mathbf{R}_{ij}^{c}-\mathbf{R}_{ij}^{f}\right)
\mathbf{P}_{ij}.
\end{equation*}

Similarly, the average risk-free rate is

\begin{equation*}
E\left[ R_{t+1}^{f}\right] =\sum_{i=1}^{2}\pi
_{i}\mathbf{R}_{ij}^{f}.
\end{equation*}

(Recall that this number does not depend on $j$.)

The unconditional distribution $\pi$ can be computed as an eigenvector of the matrix $\mathbf{P}'$ associated with eigenvalue equal to one. When $\mathbf{P}'$ has strictly positive entries, it is unique.

## Mehra and Prescott (1985) exercise

Mehra and Prescott (1985) are interested whether the economy can generate some elementary but salient features of the asset markets: a low and stable risk-free rate (about 1% in annual real terms on average), and a high risk premium (6&ndash;7% in annual terms on average).

The model is calibrated at annual frequency, with parameters $\mu =0.018$, $\delta =0.036$,
$\phi =0.43$ roughly replicating the characteristics of the consumption process.

Since the preference parameters are not directly observable, they explore the model for a wide range of preference parameters, $\beta\in(0,1)$ and $\gamma\in(0,10)$. They argue that any plausible preference parameter combinations lie comfortably within this set.

For every such combination of preference parameters, they solve the model and compute the average risk-free rate and the average risk premium, and plot all combinations of these two moments.

Function that solves for asset pricing moments in the Mehra-Prescott economy.

```python
def solve_Mehra_Prescott(model):
    
    S = model["SDF"]
    P = model["P"]
    GAMMAG = model["GAMMAG"]
    
    N = P.shape[0]
    I = np.identity(N)
    one = np.ones([N,1])
    
    # first compute the unconditional stationary distribution PII (assuming it is unique)
    # unconditional stationary distribution is the eigenvector of P' associated
    # with eigenvalue equal to 1 (the largest eigenvalue)
    eigval,eigvec = np.linalg.eig(P.transpose())
    idx = np.abs(eigval).argsort()
    eigval = eigval[idx]
    eigvec = eigvec[:,idx]
    PII = eigvec[:,-1:] /sum( eigvec[:,-1])

    # conditional gross risk-free rate (Nx1 vector)
    # Rf = 1 / (np.multiply(P,S)*one)
    Rf = 1 / ((P * S) @ one)
    # unconditional gross risk-free rate, stored as a scalar
    ERf = (Rf.transpose() @ PII).item()
    
    # recursive formula for the price-dividend ratio: q =  P*S*GAMMAG @ (q + 1)
    # where * is elementwise multiplication, @ is matrix multiplication, 1 is an Nx1 vector of ones
    # solution given by
    # q = inv(I - P*S*GAMMAG) * (P*S*GAMMAG)*1

    # solution for the infinite-horizon asset only valid
    # if P*S*G has eigenvalues inside the unit circle
    M = P*S*GAMMAG
    eigval,eigvec = np.linalg.eig(M)
    maxeig = max(eigval)

    if (maxeig < 1):
        # asset price (Nx1 vector)
        q = np.linalg.inv(I - M) @ (M @ one)
        # returns (NxN matrix of returns R(i,j)
        R = ((1/q) @ (q.transpose()+1)) * GAMMAG
        # conditional expected returns (Nx1 vector)
        EtR = (R * P) @ one
        # unconditional expected return
        ER = (EtR.transpose() @ PII).item()
        # excess returns, realized and expected
        Re = R - np.tile(Rf,[1,N])
        EtRe = EtR - Rf
        ERe = ER - ERf
    else:
        q = (np.empty([N,1]))*np.nan
        R = (np.empty([N,N]))*np.nan
        EtR = (np.empty([N,1]))*np.nan
        ER = np.nan
        Re = (np.empty([N,N]))*np.nan
        EtRe = (np.empty([N,1]))*np.nan
        ERe = np.nan

    sol = {"PII": PII, "Rf": Rf, "ERf": ERf, "q": q, "R": R, "EtR": EtR, "ER": ER, "Re": Re, "EtRe": EtRe, "ERe": ERe}
    return sol
```

Recreate Figure 4 from Mehra and Prescott (1985).

```python
# model specification
model = {"PHI": 0.43, "MU": 0.018, "DELTA": 0.036}

model["P"] = np.array([[model["PHI"], 1-model["PHI"]], [1-model["PHI"], model["PHI"]]])
model["GAMMAC"] = np.array([[1+model["MU"]+model["DELTA"],1+model["MU"]-model["DELTA"]], [1+model["MU"]+model["DELTA"],1+model["MU"]-model["DELTA"]]])
model["GAMMAG"] = model["GAMMAC"].copy()

beta_vec = np.linspace(0.001,1,200)
gamma_vec = np.linspace(0,10,400)

ERe_mat = np.zeros([len(beta_vec),len(gamma_vec)])
ERf_mat = np.zeros([len(beta_vec),len(gamma_vec)])

for i,bet in enumerate(beta_vec):
    for j,gam in enumerate(gamma_vec):
        model["BETA"] = bet
        model["GAMMA"] = gam
        model["SDF"] = model["BETA"] * np.power(model["GAMMAC"],-model["GAMMA"])
        sol = solve_Mehra_Prescott(model)
        ERe_mat[i,j] = sol["ERe"]
        ERf_mat[i,j] = sol["ERf"]
```

```python
# display results
print(f'Displaying results for parameter region beta = ({beta_vec[0]:.2f},{beta_vec[-1]:.2f}), gamma = ({gamma_vec[0]:.2f},{gamma_vec[-1]:.2f}).\n')

# replicating Figure 4
param = {'figsize' : [9,6], 'fontsize': 16, 'subplots': [1,1],
         'title': '',
         'xlim': [0,4], 'ylim': [0,2],
         'xlabel': r'average risk-free rate $E[R^f]\ (\%)$', 'ylabel': r'average risk premium $E[R^c - R^f]\ (\%)$',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': False,
         'showNBERrecessions' : False, 'showNBERrecessions_y': [0,7]}

fig,ax = ec.GenerateTSPlot(param)

ax.plot((ERf_mat.flatten()-1)*100,ERe_mat.flatten()*100,linewidth=0,marker='*',color=ec.tolColor['tolVibrantBlue'],linestyle='solid',label='admissible region');

x = ax.legend(loc="upper left")

fig.set_facecolor('#FFFFFF')
fig.savefig(graphfolder + 'asset_pricing_mehra_prescott_Rf_ER.pdf',bbox_inches='tight')
```

### Resolving the model for a wider range of preference parameters

```python
# model specification - expand parameter range
beta_vec = np.linspace(0.6,1.2,100)
gamma_vec = np.linspace(0,50,400)

ERe_mat = np.zeros([len(beta_vec),len(gamma_vec)])
ERf_mat = np.zeros([len(beta_vec),len(gamma_vec)])

for i,bet in enumerate(beta_vec):
    for j,gam in enumerate(gamma_vec):
        model["BETA"] = bet
        model["GAMMA"] = gam
        model["SDF"] = model["BETA"] * np.power(model["GAMMAC"],-model["GAMMA"])
        sol = solve_Mehra_Prescott(model)
        ERe_mat[i,j] = sol["ERe"]
        ERf_mat[i,j] = sol["ERf"]
```

```python
# display results
print(f'Displaying results for parameter region beta = ({beta_vec[0]:.2f},{beta_vec[-1]:.2f}), gamma = ({gamma_vec[0]:.2f},{gamma_vec[-1]:.2f}).\n')

# replicating Figure 4 for a wider parameter range
param = {'figsize' : [9,6], 'fontsize': 16, 'subplots': [1,1],
         'title': '',
         'xlim': [-2,10], 'ylim': [0,15],
         'xlabel': r'average risk-free rate $E[R^f]\ (\%)$', 'ylabel': r'average risk premium $E[R^c - R^f]\ (\%)$',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': False,
         'showNBERrecessions' : False, 'showNBERrecessions_y': [0,7]}

fig,ax = ec.GenerateTSPlot(param)

ax.plot((ERf_mat.flatten()-1)*100,ERe_mat.flatten()*100,linewidth=0,marker='*',color=ec.tolColor['tolVibrantBlue'],linestyle='solid',label='admissible region');

x = ax.legend(loc="upper left")

fig.set_facecolor('#FFFFFF')
fig.savefig(graphfolder + 'asset_pricing_mehra_prescott_Rf_ER_wider_parameter_range.pdf',bbox_inches='tight')
```
